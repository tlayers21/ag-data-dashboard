```python
import pandas as pd
import plotly.express as px
from pathlib import Path

processed_path = Path("../data/processed/corn_exports_2025.csv")
weekly_data = pd.read_csv(processed_path, parse_dates=["date"])
weekly_data.head()
```

```python
figure1 = px.line(
    weekly_data, 
    x="date", 
    y="weekly_exports_mt", 
    title="Weekly U.S. Corn Exports (2025)",
    markers=True,
    labels={
        "date": "Week Ending Date",
        "weekly_exports_mt": "Exports (MT)"
    }
)

figure1.show()
figure1.write_json("../data/processed/fig_weekly_exports_only.json")
```

```python
figure1.data = figure1.data[:1]

figure1.add_scatter(
    x=weekly_data["date"],
    y=weekly_data["moving_average_4w"],
    name="4-Week Moving Average",
    line=dict(color="red", width=3)
)

figure1.show()
figure1.write_json("../data/processed/fig_weekly_exports_with_4wk_avg.json")
```

```python
figure2 = px.line(
    weekly_data, 
    x="date", 
    y="std_4w", 
    title="Weekly U.S. Corn Exports (2025)",
    markers=True,
    labels={
        "date": "Week Ending Date",
        "std_4w": "Exports (MT)",
    }
)

figure2.update_traces(line=dict(color="green"))

figure2.show()
figure2.write_json("../data/processed/fig_weekly_exports_4wk_std.json")
```
