# Commodities

```python
import requests
import pandas as pd

# USDA API Key
API_KEY = "GA0LAyk7zcLgEjKMdSfOIOl7GJmL4wRleIlflcfp"

# Base URL
BASE_URL = "https://api.fas.usda.gov"

# Parameters for request
params = {"api_key": API_KEY}
ENDPOINT = f"/api/psd/commodities"

response = requests.get(BASE_URL + ENDPOINT, params=params)

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    commodities = pd.concat(
        [
            df[df["commodityName"].str.contains("corn", case=False)],
            df[df["commodityName"].str.contains("soybean", case=False)],
            df[df["commodityName"].str.contains("wheat", case=False)]
        ],
        ignore_index=True,
    )
else:
    print(f"Error: {response.status_code}, {response.text}")

commodities
```

# Countries

```python
ENDPOINT = f"/api/psd/countries"
response = requests.get(BASE_URL + ENDPOINT, params=params)

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    countries = pd.concat(
        [
            df[df["countryName"].str.contains("world", case=False)],
            df[df["countryName"].str.contains("united states", case=False)],
            df[df["countryName"].str.contains("mexico", case=False)],
            df[df["countryName"].str.contains("european", case=False)],
            df[df["countryName"].str.contains("japan", case=False)],
            df[df["countryName"].str.contains("china", case=False)],
        ],
        ignore_index=True,
    )
else:
    print(f"Error: {response.status_code}, {response.text}")

countries
```
