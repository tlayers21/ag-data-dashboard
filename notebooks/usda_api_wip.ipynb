```python
import matplotlib.pyplot as plt
import pandas as pd

# Example export data (replace this with USDA GATS data you load)
data = {
    "Year": [2020, 2021, 2022, 2023, 2024, 2025],
    "USA": [45, 50, 47, 52, 55, 53],
    "Mexico": [20, 22, 25, 27, 30, 29],
    "Brazil": [15, 18, 19, 21, 23, 22],
    "China": [30, 32, 35, 33, 38, 36]
}

df = pd.DataFrame(data)

# Function to plot exports for a given country
def plot_exports(country="USA"):
    if country not in df.columns:
        print(f"Country '{country}' not found. Available: {list(df.columns[1:])}")
        return
    
    plt.figure(figsize=(8,5))
    plt.plot(df["Year"], df[country], marker="o", label=country)
    plt.title(f"{country} Corn Exports (20202025)")
    plt.xlabel("Year")
    plt.ylabel("Exports (Million Tons)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: plot USA exports
plot_exports("USA")

# You can change to another country, e.g.
# plot_exports("Mexico")

```
